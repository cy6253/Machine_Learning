```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('titanic_train.csv')
```

```python
df1['Age'].fillna(df1['Age'].mean(), inplace = True)
df1['Cabin'].fillna('N', inplace = True)
df1['Embarked'].fillna('N', inplace = True)
```

```python
def func1(x):
  return x[0]
df1['Cabin'] = df1['Cabin'].apply(func1)
df1['Cabin'].value_counts()
```

```python
def func2(x):
  cat = ''
  if x <= -1:
    cat = 'Unknown'
  elif x <= 5:
    cat = 'Baby'
  elif x <= 12:
    cat = 'Child'
  elif x <= 18:
    cat = 'Teenager'
  elif x <= 25:
    cat = 'Student'
  elif x <= 35:
    cat = 'Young adult'
  elif x <= 60:
    cat = 'Adult'
  else:
    cat = 'Elder'
  return cat

df1['Age_group'] = df1['Age'].apply(func2)
orders = df1['Age_group'].unique()
orders
```

```python
sns.barplot(data = df1, x = 'Age_group', y = 'Survived', hue='Sex', order = orders)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

def encode_features(x):
  features = ['Cabin', 'Sex', 'Embarked']
  for i in features:
    encoder = LabelEncoder()
    encoder = encoder.fit(x[i])
    label = encoder.transform(x[i])
    x[i] = label

  return x

df1 = encode_features(df1)
df1
```

```python
data = df1.drop(['PassengerId', 'Name', 'Ticket'], axis = 1)
```

```python
X_df = data.drop(['Survived', 'Age_group'], axis = 1)
Y_df = data['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

dt_clf = DecisionTreeClassifier()
rf_clf = RandomForestClassifier()
lr_clf = LogisticRegression()
```

```python
dt_clf.fit(X_train, Y_train)
rf_clf.fit(X_train, Y_train)
lr_clf.fit(X_train, Y_train)

dtpred = dt_clf.predict(X_test)
rfpred = rf_clf.predict(X_test)
lrpred = lr_clf.predict(X_test)

acc1 = accuracy_score(dtpred, Y_test)
acc2 = accuracy_score(rfpred, Y_test)
acc3 = accuracy_score(lrpred, Y_test)

print('acc1 : ', acc1)
print('acc2 : ', acc2)
print('acc3 : ', acc3)
```

```python
from sklearn.model_selection import cross_val_score

dtscores = cross_val_score(dt_clf, X_train, Y_train, scoring='accuracy', cv = 3)
rfscores = cross_val_score(rf_clf, X_train, Y_train, scoring='accuracy', cv = 3)
lrscores = cross_val_score(lr_clf, X_train, Y_train, scoring='accuracy', cv = 3)

print('dt : ', np.mean(dtscores))
print('rf : ', np.mean(rfscores))
print('lr : ', np.mean(lrscores))
```
