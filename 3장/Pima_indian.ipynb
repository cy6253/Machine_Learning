```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, precision_recall_curve
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

df1 = pd.read_csv('diabetes.csv')
df1.info()
```

```python
def clf_eval(Y, pred):
  confusion = confusion_matrix(Y, pred)
  accuracy = accuracy_score(Y, pred)
  precision = precision_score(Y, pred)
  recall = recall_score(Y, pred)
  f1 = f1_score(Y, pred)
  print(confusion, '\n')
  print('acc : {}, precision : {}, recall = {}'.format(accuracy, precision, recall))
  print('f1 : ', f1)
```

```python
X_df = df1.iloc[:,:-1]
Y_df = df1.iloc[:, -1]

X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df, test_size = 0.2)
clf = LogisticRegression()
clf.fit(X_train, Y_train)
pred = clf.predict(X_test)
clf_eval(Y_test, pred)
```

```python
# 0값을 검사할 피처명 리스트 객체 설정
zero_features = ['Glucose', 'BloodPressure','SkinThickness','Insulin','BMI']

# 전체 데이터 건수
total_count = df1['Glucose'].count()

# 피처별로 반복 하면서 데이터 값이 0 인 데이터 건수 추출하고, 퍼센트 계산
for feature in zero_features:
    zero_count = df1[df1[feature] == 0][feature].count()
    print('{0} 0 건수는 {1}, 퍼센트는 {2:.2f} %'.format(feature, zero_count, 100*zero_count/total_count))
```

```python
df1[zero_features]=df1[zero_features].replace(0, df1[zero_features].mean()) # 0 값을 평균으로 대체
df1.describe()
```

```python
X = df1.iloc[:, :-1]
y = df1.iloc[:, -1]

# StandardScaler 클래스를 이용해 피처 데이터 세트에 일괄적으로 스케일링 적용
scaler = StandardScaler( )
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2)

# 로지스틱 회귀로 학습, 예측 및 평가 수행. 
lr_clf = LogisticRegression()
lr_clf.fit(X_train , y_train)
pred = lr_clf.predict(X_test)
clf_eval(y_test , pred)
```
